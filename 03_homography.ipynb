```python
import cv2
import numpy as np
from helper import OpenCVHelper
hlp = OpenCVHelper()

img = cv2.imread("./res/kennzeichen.png")
hlp.display_image(img)
```

```python
src = np.array([
    (52, 37), 
    (30, 153), 
    (349, 158), 
    (322, 294)
    ])

dst = np.array([
    (0, 0),
    (0, 300),
    (1100, 0),
    (1100, 300)
])

h, _ = cv2.findHomography(src, dst)
print(h)
```

```python
out = cv2.warpPerspective(img, h, (1100, 300))
hlp.display_image(out, resize_factor=0.75)
```

```python
img_audi = cv2.imread("./res/kennzeichen_audi.jpg")
hlp.display_image(img_audi, resize_factor=0.5)
```

```python
src_audi = np.array([
    (555, 291), 
    (567, 427), 
    (927, 151), 
    (930, 265)
    ])

dst_audi = np.array([
    (0, 0),
    (0, 300),
    (1100, 0),
    (1100, 300)
])

h_audi, _ = cv2.findHomography(src_audi, dst_audi)
print(h_audi)
```

```python
out_audi = cv2.warpPerspective(img_audi, h_audi, (1100, 300))
hlp.display_image(out_audi, resize_factor=0.75)
```
